```python
import numpy as np
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
from datetime import time, datetime
```

```python
data_dir = './data/price_histories/'
```

# Explorative Analysis

```python
def analyze(symbol: str):
    period = 30
    dateparser = lambda x: datetime.fromtimestamp(int(x))
    df = pd.read_csv(data_dir + f'{symbol}.csv', index_col=[0], parse_dates=['time'], date_parser=dateparser, usecols=['time', 'open', 'high', 'low', 'close', 'volumefrom', 'volumeto'])
    vol1 = df['volumefrom'].values
    vol2 = df['volumeto'].values
    vol1_conv = np.convolve(vol1, np.ones(period)/period, mode='valid')
    vol2_conv = np.convolve(vol2, np.ones(period)/period, mode='valid')
    price = df['low'].values
    price_conv = np.convolve(price, np.ones(period)/period, mode='valid')

    fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, sharex=True, figsize=(10,4))
    fig.suptitle(symbol)
    ax1.set_ylabel('Price')
    ax1.plot(price)
    ax1.plot(price_conv)
    ax2.set_ylabel(f'Vol -> USD')
    ax2.plot(vol1, color='firebrick')
    #ax2.plot(vol1_conv)
    ax3.set_ylabel(f'Vol -> {symbol}')
    ax3.plot(vol2, color='purple')
    #ax3.plot(vol2_conv)
```

```python
analyze('BTC')
analyze('ETH')
analyze('MANA')
analyze('SAND')
analyze('SHIB')
analyze('AIR')
```
